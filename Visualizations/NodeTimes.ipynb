```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dfhours = pd.read_csv("nodetimes.csv")
dfhours['Single'] = dfhours['Single'] / 3600
dfhours['Lattice'] = dfhours['Lattice'] / 3600

dfhours.set_index('Number of Nodes',inplace=True)
print(dfhours.head())
```

```python
%matplotlib inline
dfhours.plot(figsize=(10,6))
plt.ylabel("Runtime (Hours)")
plt.title("Single and Lattice Filter Runtime")
plt.savefig("nodetimes.pdf")
```
