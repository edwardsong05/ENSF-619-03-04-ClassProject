```python
from mpl_toolkits.basemap import Basemap
import matplotlib.pyplot as plt
import numpy as np
import json
```

```python
#Read in labeled tweets
jlines = []
with open("../Data/CombinedLabeled.json") as f:
    for line in f:
        jlines.append(json.loads(line))
print("\nFile closed")
```

```python
##Read in coordinates
relGeoLines = []
for jline in jlines:
    if jline['isRelevant']:
        if jline['geo'] is not None:
            relGeoLines.append(jline)
```

```python
##parse coordinates
longLats = []
for line in relGeoLines:
    longLats.append(line['geo']['coordinates'])
longLats = np.array(longLats)
```

```python
##plot tweets on map
%matplotlib inline
plt.figure()

m=Basemap(llcrnrlon=-180, llcrnrlat=-65,urcrnrlon=180,urcrnrlat=80)
m.drawmapboundary(fill_color='#A6CAE0', linewidth=0)
m.fillcontinents(color='#2a6619', alpha=0.3)

m.scatter(longLats[:,1], longLats[:,0], c = '#d85f08')
plt.tight_layout()
```
